### Example 2.1 Chicken price over time 

```python
# import all data from package astsadata 
chicken = pd.DataFrame(astsadata.chicken)
```

```python
def change_time_value(t):
    
    """
    For chicken price data, change the Time YYYY-mm to value as the 
    predictor for chicken price 
    
    Value = YYYY + mm / 12
    """

    result = float(t.strftime('%Y-%m').split('-')[0]) + \
        (float(t.strftime('%Y-%m').split('-')[1]) - 1) / 12
    return result

time = pd.Series([change_time_value(idx) for idx in chicken.index], index=chicken.index)
```

```python
# linear regression price ~ time AR(1)
data_reg = pd.concat([time, chicken], axis=1)
data_reg.columns = ["time", "chicken"]

fit = ols(formula="chicken ~ time", data=data_reg).fit()
print(fit.summary())
```

```python
# plot time series and fitted line 
ax = chicken.plot(xlabel="Time", ylabel="cents per pound", legend=False, figsize=(8,3))
ax.plot(data_reg.index, fit.params.Intercept + fit.params.time * time, color="orange")
plt.xlabel("Time", fontsize=10)
plt.ylabel("US dollar", fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title("Monthly price of Chicken (per pound)", fontsize=15)
plt.legend(fontsize=10);
plt.show()
```

### Example 2.2 Mortality ~ Tempurature + Pollution

- what is t?
- anova interpretation
- Math expression for SSE, DF, MSE, AIC, BIC

```python
cmort = pd.DataFrame(astsadata.cmort)
tempr = pd.DataFrame(astsadata.tempr)
part = pd.DataFrame(astsadata.part)
```

```python
fig, axes = plt.subplots(nrows=3, figsize=(9, 12))

cmort.plot(ax=axes[0], xlabel="Time", title="Cardiovascular Mortality", legend=False)
tempr.plot(ax=axes[1], xlabel="Time", title="Temperature", legend=False)
part.plot(ax=axes[2], xlabel="Time", title="Particulates", legend=False)

fig.tight_layout()
plt.show()
```

```python
# check pairwise relationship 
pairs = pd.concat([cmort, tempr, part], axis=1)
pairs.columns = ["cmort", "tempr", "part"]

pd.plotting.scatter_matrix(pairs, figsize=(7, 5))
plt.show() 
```

#### Insights
- Temperature is independent from pollution
- Both mortality and part is right skewed
- Mortality is negatively correlated with temperature
- Mortality is positively correlated with pollution 

```python
# create different predictors 
trend = pd.Series([idx.ordinal for idx in cmort.index], index=cmort.index) # time
temp = tempr - tempr.mean() # center temperature
temp2 = temp.pow(2) # square it

mortality_reg = pd.concat([cmort, temp, temp2, trend, part], axis=1)
mortality_reg.columns = ["cmort", "temp", "temp2", "trend", "part"]

m218 = ols('cmort ~ trend', data=mortality_reg).fit()
m219 = ols('cmort ~ trend + temp', data=mortality_reg).fit()
m220 = ols('cmort ~ trend + temp + temp2', data=mortality_reg).fit()
m221 = ols('cmort ~ trend + temp + temp2 + part', data=mortality_reg).fit()
anovaResults = anova_lm(m218, m219,m220, m221)
print(anovaResults)
```

### Example 2.3 Recuritment series ~ South Oscillation Index 
- dynamic time series in python pydlm

```python
rec = pd.DataFrame(astsadata.rec)
soi = pd.DataFrame(astsadata.soi)
```

```python
# Regular regression 
Recu_reg = pd.concat([rec, soi.shift(6)], axis=1).dropna()
Recu_reg.columns = ["rec", "soiL6"]
fit = ols(formula="rec ~ soiL6", data=Recu_reg).fit()
print(fit.summary())
```

```python
Recu_reg["rec"].plot(xlabel="Time", figsize = (15,5))
fit.fittedvalues.plot(color='orange')
plt.xlabel("Time", fontsize=15)
plt.ylabel("Number of new fish", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title("Number of new fish", fontsize=20)
plt.show()
```
